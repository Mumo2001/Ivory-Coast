Cells_geo_description dataset describes the column names of the cells-geo dataset and their formats.

```
#load telcom datasets 1,2 and 3
df_tel_dataset1 = pd.read_csv("Telcom_dataset.csv")
df_tel_dataset1
```

```
df_tel_dataset2 = pd.read_csv("Telcom_dataset2.csv")
df_tel_dataset2
```

```
df_tel_dataset3 = pd.read_csv("Telcom_dataset3.csv")
df_tel_dataset3
```

The Telcom datasets are the CDR data records from Telcom in a period of 3 days.

```
#Clean data in the Telcom data sets
#some of the column names have an error and some are not alike
#we will the column names that have an error.
```

```
df_tel_dataset1.rename(columns = {'PRODUTC' : 'PRODUCT', 'DATETIME' : 'DATE_TIME'}, inplace = True)
df_tel_dataset1.columns
```

```
df_tel_dataset2.rename(columns = {'DW_A_NUMBER' : 'DW_A_NUMER_INT', 'DW_B_NUMBER' : 'DW_B_NUMBER_INT'}, inplace= True)
df_tel_dataset2.columns
```

```
df_tel_dataset3.rename(columns = {'CELLID' : 'CELL_ID','SIET_ID': 'SITE_ID'},inplace= True)
df_tel_dataset3.columns
```

```
df_cells_geo.duplicated().sum()
```

```
df_cells_geo.isnull()
```

```
df_cells_geo = df_cells_geo.dropna(axis=1)
df_cells_geo
```

```
#merge data frames
#telcom datasets.
df_4 = pd.concat([df_tel_dataset1, df_tel_dataset2, df_tel_dataset3],axis = 0)
df_4.head()
```

```
df_4.duplicated().sum()
```

```
df_4 = df_4.drop_duplicates()
```

```
df_5 = df_4.merge(df_cells_geo,left_on = "SITE_ID" ,right_on = "SITE_CODE")
df_5
```

Here we merged 4 datasets; the telecom datasets and the cells_geo dataset.

```
#Explore and sort the data
#group the the product and villies(cities) columns
df_5.groupby(['PRODUCT'])['VILLES'].describe()
```

```
#sort the rows where the columns has certain values
#here we will consider the Product column
values = ['Voice']
df_5[df_5.PRODUCT.isin(values)]
```

```
values = ['sms']
df_5[df_5.PRODUCT.isin(values)]
```

```
#Spilt the 'DATE_TIME' column to time and date
df_5[['date', 'time']] = df_5['DATE_TIME'].str.split(' ',expand = True)
```

```
df_5
```

```
#Which ones were the most used city for the three days?
df_5['VILLES'].value_counts()
```

```
#Which cities were the most used during business and home hours?
df_5.time[(df_5['time']>='0:08:00.0') & (df_5['time']<='0:17:00.0')].count()
```

```
#home hours.
df_5.time[(df_5['time']<='0:08:00.0') | (df_5['time']>='0:17:00.0')].count()
```
